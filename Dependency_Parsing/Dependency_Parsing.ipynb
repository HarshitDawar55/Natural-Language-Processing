# Dependency_Parsing by `Mr. Harshit Dawar!`
* This is the process of checking how dependent the words are in a sentence!

```python
import spacy
english_model_nlp = spacy.load("en_core_web_sm")
```

```python
# Creating a sentence for the demo!
sentence = english_model_nlp("This is my first class for NLP in India \
by Mr. Harshit Dawar!")
```

```python
# Drawing a diagram that showcases the dependency of words on the other words!
# "dep" in style tells that dependency graph is required
# "distance" keyword tells the total length of the diagram
spacy.displacy.render(sentence, style = "dep", options = {"distance": 75})
```
